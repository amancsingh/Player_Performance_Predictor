# Player Performance Predictor - Phase 2: Model Training (Definitive Strategy)

**Goal:** Combine a data-driven filtering strategy (n_min threshold) with an optimized model configuration (max_depth, max_features) to build a robust, accurate, and fast-training prediction engine.

### Step 1: Import Libraries & Load Data

```python
import pandas as pd
import sqlite3
import joblib
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# Set plot style
sns.set_style('whitegrid')

# Connect to the database and load the data
db_path = '../cricket_data.db'
conn = sqlite3.connect(db_path)

batting_df = pd.read_sql_query("SELECT * FROM batting_innings", conn)
bowling_df = pd.read_sql_query("SELECT * FROM bowling_innings", conn)

conn.close()

# Convert date column to a proper datetime format for sorting
batting_df['date'] = pd.to_datetime(batting_df['date'])
bowling_df['date'] = pd.to_datetime(bowling_df['date'])

print("Data loaded and date column converted successfully.")
```

## Part A: Training the Batting Prediction Model

### Step 2: Batting Coverage Analysis to Determine `n_min`

```python
# Calculate the number of innings batted by each player
player_counts_batting = batting_df['player'].value_counts()

# Calculate cumulative coverage for batting
cumulative_coverage_batting = player_counts_batting.cumsum() / len(batting_df) * 100

# Plot the coverage for batting
plt.figure(figsize=(12, 7))
plt.plot(np.arange(1, len(cumulative_coverage_batting) + 1), cumulative_coverage_batting.values)
plt.title('Batting: Cumulative Coverage of Innings by Number of Players', fontsize=16)
plt.xlabel('Number of Players (sorted by frequency)', fontsize=12)
plt.ylabel('Percentage of Total Innings Covered (%)', fontsize=12)
plt.grid(True, which='both', linestyle='--')

target_coverage = 85 # Setting a practical target to find the 'knee point'
players_for_target_coverage = np.searchsorted(cumulative_coverage_batting, target_coverage) + 1

plt.axhline(y=target_coverage, color='g', linestyle='--', label=f'{target_coverage}% Coverage')
plt.axvline(x=players_for_target_coverage, color='g', linestyle='--')
print(f"To cover {target_coverage}% of all batting innings, we need to include the top {players_for_target_coverage} most frequent players.")

plt.legend()
plt.show()
```

### Step 3: Feature Engineering & Reduction for Batting Data

```python
# --- Create Contextual & Form Features --- #
print("Creating features for batting data...")
match_teams = batting_df.groupby('match_id')['team'].unique().apply(list).to_dict()
def find_opposition(row):
    teams_in_match = match_teams.get(row['match_id'])
    if teams_in_match and len(teams_in_match) == 2:
        return teams_in_match[1] if teams_in_match[0] == row['team'] else teams_in_match[0]
    return 'N/A'
batting_df['against_team'] = batting_df.apply(find_opposition, axis=1)
batting_df = batting_df.sort_values(by=['player', 'date'])
batting_df['form_last_5_innings'] = batting_df.groupby('player')['runs'].transform(lambda x: x.rolling(window=5, min_periods=1).mean().shift(1))
batting_df['form_last_5_innings'] = batting_df['form_last_5_innings'].fillna(0)

# --- Data-Driven Feature Reduction --- #
print("Performing feature reduction based on our coverage analysis...")
# Get the list of players that meet our coverage target
frequent_batters = player_counts_batting.head(players_for_target_coverage).index

# For venues and teams, we can use a simpler fixed threshold as their numbers are smaller
frequent_venues = batting_df['venue'].value_counts().head(200).index
frequent_teams = batting_df['against_team'].value_counts().head(50).index

print(f"Final number of frequent batters being kept: {len(frequent_batters)}")

batting_df['player_reduced'] = batting_df['player'].where(batting_df['player'].isin(frequent_batters), 'Other_Player')
batting_df['venue_reduced'] = batting_df['venue'].where(batting_df['venue'].isin(frequent_venues), 'Other_Venue')
batting_df['against_team_reduced'] = batting_df['against_team'].where(batting_df['against_team'].isin(frequent_teams), 'Other_Team')

print("Feature engineering and reduction complete for batting data.")
```

### Step 4: Prepare Batting Data for Modeling

```python
categorical_features_reduced = ['player_reduced', 'venue_reduced', 'against_team_reduced']
numerical_features = ['form_last_5_innings']
target_batting = 'runs'

encoded_features = pd.get_dummies(batting_df[categorical_features_reduced], columns=categorical_features_reduced)
X_batting = pd.concat([encoded_features, batting_df[numerical_features]], axis=1)
y_batting = batting_df[target_batting]

print("Batting data prepared.")
print("Final optimized shape of feature matrix (X):", X_batting.shape)
```

### Step 5: Train the Batting Model

```python
X_train, X_test, y_train, y_test = train_test_split(X_batting, y_batting, test_size=0.2, random_state=42)

# --- MODIFIED: Removed max_features to allow the model to be more detailed ---
batting_model = RandomForestRegressor(
    n_estimators=100,
    max_depth=20,
    random_state=42,
    n_jobs=-1,
    verbose=2
)

print("Training the batting model with more detail...")
batting_model.fit(X_train, y_train)
print("Model training complete.")

predictions = batting_model.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print(f"\n--- Batting Model Evaluation ---")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R-squared (R2 Score): {r2:.2f}")
```

### Step 6: Save the Trained Batting Model

```python
if not os.path.exists('../models'):
    os.makedirs('../models')

joblib.dump(batting_model, '../models/batting_predictor_model.joblib')
joblib.dump(X_batting.columns, '../models/batting_model_columns.joblib')

print(f"Batting model and columns saved successfully.")
```

## Part B: Training the Bowling Prediction Model

### Step 7: Bowling Coverage Analysis to Determine `n_min`

```python
# Calculate the number of innings bowled by each player
player_counts_bowling = bowling_df['player'].value_counts()

# Calculate cumulative coverage for bowling
cumulative_coverage_bowling = player_counts_bowling.cumsum() / len(bowling_df) * 100

# Plot the coverage for bowling
plt.figure(figsize=(12, 7))
plt.plot(np.arange(1, len(cumulative_coverage_bowling) + 1), cumulative_coverage_bowling.values)
plt.title('Bowling: Cumulative Coverage of Innings by Number of Players', fontsize=16)
plt.xlabel('Number of Players (sorted by frequency)', fontsize=12)
plt.ylabel('Percentage of Total Innings Covered (%)', fontsize=12)
plt.grid(True, which='both', linestyle='--')

target_coverage_bowling = 85
players_for_target_coverage_bowling = np.searchsorted(cumulative_coverage_bowling, target_coverage_bowling) + 1

plt.axhline(y=target_coverage_bowling, color='g', linestyle='--', label=f'{target_coverage_bowling}% Coverage')
plt.axvline(x=players_for_target_coverage_bowling, color='g', linestyle='--')
print(f"To cover {target_coverage_bowling}% of all bowling innings, we need to include the top {players_for_target_coverage_bowling} most frequent players.")

plt.legend()
plt.show()
```

### Step 8: Feature Engineering & Reduction for Bowling Data

```python
print("\nCreating features for bowling data...")
bowling_df['against_team'] = bowling_df.apply(find_opposition, axis=1)
bowling_df = bowling_df.sort_values(by=['player', 'date'])
bowling_df['form_last_5_wickets'] = bowling_df.groupby('player')['wickets'].transform(lambda x: x.rolling(window=5, min_periods=1).mean().shift(1))
bowling_df['form_last_5_wickets'] = bowling_df['form_last_5_wickets'].fillna(0)

print("Performing feature reduction for bowling data...")
frequent_bowlers = player_counts_bowling.head(players_for_target_coverage_bowling).index
print(f"Final number of frequent bowlers being kept: {len(frequent_bowlers)}")

bowling_df['player_reduced'] = bowling_df['player'].where(bowling_df['player'].isin(frequent_bowlers), 'Other_Player')
bowling_df['venue_reduced'] = bowling_df['venue'].where(bowling_df['venue'].isin(frequent_venues), 'Other_Venue')
bowling_df['against_team_reduced'] = bowling_df['against_team'].where(bowling_df['against_team'].isin(frequent_teams), 'Other_Team')

print("Feature engineering and reduction complete for bowling data.")
```

### Step 9: Prepare Bowling Data for Modeling

```python
categorical_features_reduced = ['player_reduced', 'venue_reduced', 'against_team_reduced']
numerical_features = ['form_last_5_wickets']
target_bowling = 'wickets'

encoded_features_bowling = pd.get_dummies(bowling_df[categorical_features_reduced], columns=categorical_features_reduced)
X_bowling = pd.concat([encoded_features_bowling, bowling_df[numerical_features]], axis=1)
y_bowling = bowling_df[target_bowling]

print("Bowling data prepared.")
print("Final optimized shape of feature matrix (X):", X_bowling.shape)
```

### Step 10: Train the Bowling Model

```python
X_train, X_test, y_train, y_test = train_test_split(X_bowling, y_bowling, test_size=0.2, random_state=42)

bowling_model = RandomForestRegressor(
    n_estimators=100,
    max_depth=20,
    max_features='sqrt',
    random_state=42,
    n_jobs=-1,
    verbose=2
)
print("Training the bowling model with optimized parameters...")
bowling_model.fit(X_train, y_train)
print("Model training complete.")

predictions = bowling_model.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print(f"\n--- Bowling Model Evaluation ---")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R-squared (R2 Score): {r2:.2f}")
```

### Step 11: Save the Trained Bowling Model

```python
joblib.dump(bowling_model, '../models/bowling_predictor_model.joblib')
joblib.dump(X_bowling.columns, '../models/bowling_model_columns.joblib')

print(f"Bowling model and columns saved successfully.")
```

```python
### Part C: Real-World Prediction Test (Corrected & Robust)

import joblib

# Load the saved model and, crucially, the list of column names it was trained on
saved_batting_model = joblib.load('../models/batting_predictor_model.joblib')
saved_batting_columns = joblib.load('../models/batting_model_columns.joblib')

# Also load the bowling model for a complete test
saved_bowling_model = joblib.load('../models/bowling_predictor_model.joblib')
saved_bowling_columns = joblib.load('../models/bowling_model_columns.joblib')


def predict_performance(player_name, venue, opposition, full_batting_df, full_bowling_df):
    """
    Predicts both runs and wickets for a player given the context.
    """
    
    # --- Batting Prediction ---
    try:
        # 1. Calculate player's current batting form from the original, complete dataframe
        player_batting_history = full_batting_df[full_batting_df['player'] == player_name].sort_values(by='date')
        if player_batting_history.empty:
            predicted_runs = "Batting data not found."
        else:
            last_5_innings_form = player_batting_history['runs'].tail(5).mean()

            # 2. Create the input template with the exact columns the model expects
            batting_input = pd.DataFrame(columns=saved_batting_columns, index=[0]).fillna(0)

            # 3. Construct the column names from the input
            player_col = f"player_reduced_{player_name}"
            venue_col = f"venue_reduced_{venue}"
            opposition_col = f"against_team_reduced_{opposition}"

            # 4. Populate the input template
            if player_col in batting_input.columns:
                batting_input.loc[0, player_col] = 1
            else:
                batting_input.loc[0, 'player_reduced_Other_Player'] = 1

            if venue_col in batting_input.columns:
                batting_input.loc[0, venue_col] = 1
            else:
                batting_input.loc[0, 'venue_reduced_Other_Venue'] = 1
            
            if opposition_col in batting_input.columns:
                batting_input.loc[0, opposition_col] = 1
            else:
                batting_input.loc[0, 'against_team_reduced_Other_Team'] = 1

            batting_input['form_last_5_innings'] = last_5_innings_form
            
            # 5. Make the prediction
            prediction = saved_batting_model.predict(batting_input)
            predicted_runs = f"{prediction[0]:.2f}"
    except Exception as e:
        predicted_runs = f"An error occurred during batting prediction: {e}"

    # --- Bowling Prediction ---
    try:
        # 1. Calculate player's current bowling form
        player_bowling_history = full_bowling_df[full_bowling_df['player'] == player_name].sort_values(by='date')
        if player_bowling_history.empty:
            predicted_wickets = "No bowling history found for this player."
        else:
            last_5_wickets_form = player_bowling_history['wickets'].tail(5).mean()

            # 2. Create the input template
            bowling_input = pd.DataFrame(columns=saved_bowling_columns, index=[0]).fillna(0)
            
            # 3. Construct column names
            player_col_bowl = f"player_reduced_{player_name}"
            venue_col_bowl = f"venue_reduced_{venue}"
            opposition_col_bowl = f"against_team_reduced_{opposition}"

            # 4. Populate the template
            if player_col_bowl in bowling_input.columns:
                bowling_input.loc[0, player_col_bowl] = 1
            else:
                bowling_input.loc[0, 'player_reduced_Other_Player'] = 1

            if venue_col_bowl in bowling_input.columns:
                bowling_input.loc[0, venue_col_bowl] = 1
            else:
                bowling_input.loc[0, 'venue_reduced_Other_Venue'] = 1

            if opposition_col_bowl in bowling_input.columns:
                bowling_input.loc[0, opposition_col_bowl] = 1
            else:
                bowling_input.loc[0, 'against_team_reduced_Other_Team'] = 1

            bowling_input['form_last_5_wickets'] = last_5_wickets_form
            
            # 5. Make the prediction
            prediction_wickets = saved_bowling_model.predict(bowling_input)
            predicted_wickets = f"{prediction_wickets[0]:.2f}"
    except Exception as e:
        predicted_wickets = f"An error occurred during bowling prediction: {e}"


    return predicted_runs, predicted_wickets


# --- Let's test it! ---
# IMPORTANT: Use the exact names from your database. Let's find a player who is both a batter and bowler.
# For this test, you might need to find a prominent all-rounder from your data.
# Let's try with a placeholder name. Replace with a real name from your `frequent_batters` and `frequent_bowlers` lists.
player = "RA Jadeja" 
venue = "MA Chidambaram Stadium, Chepauk, Chennai"
opposition = "Australia"

predicted_score, predicted_wickets = predict_performance(player, venue, opposition, batting_df, bowling_df)

print(f"--- Prediction Result ---")
print(f"Player: {player}")
print(f"Venue: {venue}")
print(f"Opposition: {opposition}")
print(f"\nPredicted Runs: {predicted_score}")
print(f"Predicted Wickets: {predicted_wickets}")
```
